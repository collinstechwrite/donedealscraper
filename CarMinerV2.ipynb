```python

from openpyxl import Workbook, load_workbook



    
    
import xlsxwriter
workbook = xlsxwriter.Workbook('analysis.xlsx')
worksheet = workbook.add_worksheet()
workbook.close()


    
    

Row_Incrementer = 2

def spreadsheet_titles():
    Row_Incrementer = 1
    workbook = load_workbook(filename="analysis.xlsx")
    sheet = workbook.active

    sheet["A" + str(Row_Incrementer)] = "TITLE"
    sheet["B" + str(Row_Incrementer)] = "URL"
    sheet["C" + str(Row_Incrementer)] = "Location"
    sheet["D" + str(Row_Incrementer)] = "Price"
    sheet["E" + str(Row_Incrementer)] = "Make"
    sheet["F" + str(Row_Incrementer)] = "Model"
    sheet["G" + str(Row_Incrementer)] = "Year"
    sheet["H" + str(Row_Incrementer)] = "Fuel Type"
    sheet["I" + str(Row_Incrementer)] = "Body Type"
    sheet["J" + str(Row_Incrementer)] = "Engine Size (Litres)"
    sheet["K" + str(Row_Incrementer)] = "NCT Expiry"
    sheet["L" + str(Row_Incrementer)] = "Country of Registration"
    sheet["M" + str(Row_Incrementer)] = "Doors"
    sheet["N" + str(Row_Incrementer)] = "Mileage"
    sheet["O" + str(Row_Incrementer)] = "Transmission"
    sheet["P" + str(Row_Incrementer)] = "Seats"
    sheet["Q" + str(Row_Incrementer)] = "Road Tax"
    sheet["R" + str(Row_Incrementer)] = "Previous Owners"
    sheet["S" + str(Row_Incrementer)] = "Colour"
    sheet["T" + str(Row_Incrementer)] = "TOWN"
    sheet["U" + str(Row_Incrementer)] = "SELLER"
    sheet["V" + str(Row_Incrementer)] = "Dealer"

    workbook.save(filename="analysis.xlsx")   
    
        
    
def row_increment(Title,
                  URL,
                  Location,
                  Price,
                  Make,
                  Model,
                  Year,
                  FuelType,
                  BodyType,
                  EngineSize,
                  NCTExpiry,
                  Country,
                  Doors,
                  Mileage,
                  Transmission,
                  Seats,
                  RoadTax,
                  Owners,
                  Colour,
                  TOWN,
                  SELLER,
                  Dealer):

    global Row_Incrementer
    Row_Incrementer += 1


    workbook = load_workbook(filename="analysis.xlsx")
    sheet = workbook.active

    sheet["A" + str(Row_Incrementer)] = Title
    sheet["B" + str(Row_Incrementer)] = URL
    sheet["C" + str(Row_Incrementer)] = Location
    sheet["D" + str(Row_Incrementer)] = Price    
    sheet["E" + str(Row_Incrementer)] = Make
    sheet["F" + str(Row_Incrementer)] = Model
    sheet["G" + str(Row_Incrementer)] = Year
    sheet["H" + str(Row_Incrementer)] = FuelType
    sheet["I" + str(Row_Incrementer)] = BodyType
    sheet["J" + str(Row_Incrementer)] = EngineSize
    sheet["K" + str(Row_Incrementer)] = NCTExpiry
    sheet["L" + str(Row_Incrementer)] = Country
    sheet["M" + str(Row_Incrementer)] = Doors
    sheet["N" + str(Row_Incrementer)] = Mileage
    sheet["O" + str(Row_Incrementer)] = Transmission
    sheet["P" + str(Row_Incrementer)] = Seats
    sheet["Q" + str(Row_Incrementer)] = RoadTax
    sheet["R" + str(Row_Incrementer)] = Owners
    sheet["S" + str(Row_Incrementer)] = Colour
    sheet["T" + str(Row_Incrementer)] = TOWN
    sheet["U" + str(Row_Incrementer)] = SELLER
    sheet["V" + str(Row_Incrementer)] = Dealer # originally E

    
    
    
    workbook.save(filename="analysis.xlsx")    
    
    
    
spreadsheet_titles()


# importing re module
import requests
import re

count = 28


def get_LOCATION_url(url):
    
    try:

        tag5 = 'content="For Sale in '
        tag6 = ':'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""
    
"""------------------------------------------------------------------"""
    
def get_Make_url(url):
    
    try:

        tag5 = '"name":"make","value":"'
        tag6 = '","displayName":"Make / Model"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
    
def get_Model_url(url):
    
    try:


        tag5 = '{"name":"model","value":"'
        tag6 = '","displayName":"Model"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""


def get_Year_url(url):
    
    try:


        tag5 = '{"name":"year","value":"'
        tag6 = '","displayName":"Year"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_FuelType_url(url):
    
    try:

        tag5 = '{"name":"fuelType","value":"'
        tag6 = '","displayName":"Fuel Type"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_BodyType_url(url):
    
    try:

        tag5 = '{"name":"bodyType","value":"'
        tag6 = '","displayName":"Body Type"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""


def get_EngineSize_url(url):
    
    try:


        tag5 = '{"name":"engine","value":"'
        tag6 = '","displayName":"Engine Size (Litres)"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""


def get_NCTExpiry_url(url):
    
    try:


        tag5 = '{"name":"NCT","value":"'
        tag6 = '","displayName":"NCT Expiry"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""



def get_Country_url(url):
    
    try:

        tag5 = '{"name":"country","value":"'
        tag6 = '","displayName":"Country of Registration"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    

def get_Doors_url(url):
    
    try:


        tag5 = '{"name":"numDoors","value":"'
        tag6 = '","displayName":"Doors"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    

def get_Mileage_url(url):
    
    try:


        tag5 = '{"name":"mileage","value":"'
        tag6 = '","displayName":"Mileage"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_Transmission_url(url):
    
    try:


        tag5 = '{"name":"transmission","value":"'
        tag6 = '","displayName":"Transmission"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_Seats_url(url):
    
    try:


        tag5 = '{"name":"seats","value":"'
        tag6 = '","displayName":"Seats","badge":'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_RoadTax_url(url):
    
    try:


        tag5 = '{"name":"roadTax","value":"'
        tag6 = '","displayName":"Road Tax"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
    
def get_Owners_url(url):
    
    try:


        tag5 = 'name":"previousOwners","value":"'
        tag6 = '","displayName":"Previous Owners"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_Colour_url(url):
    
    try:


        tag5 = '{"name":"colour","value":"'
        tag6 = '","displayName":"Colour"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_TOWN_url(url):
    
    try:


        tag5 = '"countyTown":"'
        tag6 = '"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    
def get_SELLER_url(url):
    
    try:


        tag5 = '"publisherName":"'
        tag6 = '","publisherPhoneEnc"'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    
    except:
        return ""

    

"""------------------------------------------------------------------""" 
    
def get_Price_url(url):
    
    try:


        tag5 = '<meta property="product:price:amount" content="'
        tag6 = '" />'
        reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
        res4 = re.findall(reg_str3, url)[0]
        
        return res4
    except:
        return ""
     

        
def get_Dealer_url(url):    
        try:

            tag5 = '<p class="card__header-title" data-reactid="8">'
            tag6 = '</p><ul class="card__header-subinfo" data-reactid="9">'
            reg_str3 = "" + tag5 + "(.*?)" + tag6 + ""
            res4 = re.findall(reg_str3, url)[0]

            return res4
        
        except:
            
            return "SELLER NOT LISTED"

            
        
        
        
        

def findcars():
    global count
    r = requests.get('https://www.donedeal.ie/cars?make=Volkswagen;model:Golf,Passat,Polo,Tiguan,Jetta&start=' + str(count))
    #print(r.text) 
    test_str = str(r.text)


    # initializing tag
    tag1 = 'g" title="'
    tag2 ='"'

    tag3 = '><a href="'
    tag4 = '"'




    # regex to extract required strings
    reg_str = "" + tag1 + "(.*?)" + tag2 + ""
    reg_str2 = "" + tag3 + "(.*?)" + tag4 + ""

    res = re.findall(reg_str, test_str)
    res2 = re.findall(reg_str2, test_str)

    res3 = []

    for i in res2:
        if 'cars-for-sale' in i:
            res3.append(i)


    # printing result
    for i,z in zip(res,res3):
        print(i)
        print(z)
        
        
        process_string = requests.get(z)
        processing_string = process_string.text

        Location = get_LOCATION_url(processing_string)
        Price = get_Price_url(processing_string)
        Make = get_Make_url(processing_string)
        Model = get_Model_url(processing_string)
        Year = get_Year_url(processing_string)
        FuelType = get_FuelType_url(processing_string)
        BodyType = get_BodyType_url(processing_string)
        EngineSize = get_EngineSize_url(processing_string)
        NCTExpiry = get_NCTExpiry_url(processing_string)
        Country = get_Country_url(processing_string)
        Doors = get_Doors_url(processing_string)
        Mileage = get_Mileage_url(processing_string)
        Transmission = get_Transmission_url(processing_string)
        Seats = get_Seats_url(processing_string)
        RoadTax = get_RoadTax_url(processing_string)
        Owners = get_Owners_url(processing_string)
        Colour = get_Colour_url(processing_string)
        TOWN = get_TOWN_url(processing_string)
        SELLER = get_SELLER_url(processing_string)
        Dealer = get_Dealer_url(processing_string)
        
        
        print(Location)
        print(Price)
        print(Dealer)
              
        row_increment(
            i,
            z,
            Location,
            Price,          
            Make,
            Model,
            Year,
            FuelType,
            BodyType,
            EngineSize,
            NCTExpiry,
            Country,
            Doors,
            Mileage,
            Transmission,
            Seats,
            RoadTax,
            Owners,
            Colour,
            TOWN,
            SELLER,
            Dealer)
        
        #spreadsheet_update()
        print("----------------------------------------------------------------------------------")
    
    
    
    count += 28
    
while count < 281:
    findcars()
    print('FINDING NEW BATCH', count)
    
    
```
